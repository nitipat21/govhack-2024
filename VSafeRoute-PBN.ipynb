```python
import pandas as pd
df = pd.read_csv('Principal_Bicycle_Network_(PBN) (1).csv')
df.head()
```

**The percentage of existing on road route has strategic cycling corridor**

```python
filtered_df = df[
    (df['type'] == 'On Road') &
    (df['status'] == 'Existing') &
    (df['strategic_cycling_corridor'] == 'Yes')
]
count_on_exist_yes = len(filtered_df)
print("Number of Exisitng On Road has Strategic Cycling Corridor:", count_on_exist_yes)
```

```python
filtered_df = df[
    (df['type'] == 'On Road') &
    (df['status'] == 'Existing')
]
count_on_exist = len(filtered_df)
print("Number of the overal Exisitng On Road:", count_on_exist)
```

```python
percentage_on_exist = count_on_exist_yes /  count_on_exist  * 100
print("Percentage of Existing On Road has Strategic Cycling Corridor:", percentage_on_exist, "%")
```

**The percentage of proposed on road route has strategic cycling corridor**

```python
filtered_df = df[
    (df['type'] == 'On Road') &
    (df['status'] == 'Proposed') &
    (df['strategic_cycling_corridor'] == 'Yes')
]
filtered_df.head()
count_on_prop_yes = len(filtered_df)
print("Number of Proposed On Road has Strategic Cycling Corridor:", count_on_prop_yes)
```

```python
filtered_df = df[
    (df['type'] == 'On Road') &
    (df['status'] == 'Proposed')
]
count_on_prop = len(filtered_df)
print("Number of the overall Proposed On Road:", count_on_prop)
```

```python
percentage_on_prop = (count_on_prop_yes / count_on_prop) * 100
print("Percentage of Proposed On Road has Strategic Cycling Corridor:", percentage_on_prop, "%")
```

**The percentage of existing off road route has strategic cycling corridor**

```python
filtered_df = df[
    (df['type'] == 'Off Road') &
    (df['status'] == 'Existing') &
    (df['strategic_cycling_corridor'] == 'Yes')
]
filtered_df.head()
count_off_exist_yes = len(filtered_df)
print("Number of Exisitng Off Road has Strategic Cycling Corridor:", count_off_exist_yes)
```

```python
filtered_df = df[
    (df['type'] == 'Off Road') &
    (df['status'] == 'Existing')
]
filtered_df.head()
count_off_exist = len(filtered_df)
print("Number of the Overall Exisitng Off Road", count_off_exist)
```

```python
percentage_off_exist = (count_off_exist_yes / count_off_exist) * 100
print("Percentage of Existing Off Road has Strategic Cycling Corridor:", percentage_off_exist, "%")
```

**The percentage of proposed off road route has strategic cycling corridor**

```python
filtered_df = df[
    (df['type'] == 'Off Road') &
    (df['status'] == 'Proposed') &
    (df['strategic_cycling_corridor'] == 'Yes')
]
filtered_df.head()
count_off_prop_yes = len(filtered_df)
print("Number of Proposed On Road has Strategic Cycling Corridor:", count_off_prop_yes)
```

```python
filtered_df = df[
    (df['type'] == 'Off Road') &
    (df['status'] == 'Proposed')
]
filtered_df.head()
count_off_prop = len(filtered_df)
print("Number of the Overall Proposed Off Road:", count_off_prop)
```

```python
percentage_off_prop = (count_off_prop_yes / count_off_prop) * 100
print("Percentage of Proposed Off Road has Strategic Cycling Corridor:", percentage_off_prop, "%")
```

**Combine into a table for Strategic Cycling Corridor**

```python
scc_df = pd.DataFrame({
    'Type': ['On Road', 'Off Road'],
    'Existing': [count_on_exist_yes, count_off_exist_yes],
    'Proposed': [count_on_prop_yes, count_off_prop_yes],
    'Percentage Existing (%)': [percentage_on_exist, percentage_off_exist],
    'Percentage Proposed (%)': [percentage_on_prop, percentage_off_prop]
})
scc_df
```
